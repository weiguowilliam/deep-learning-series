```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def get_data():
    dt = pd.read_csv('ecommerce_data.csv')
    dt = dt.as_matrix()
    N,D = dt.shape
    X = dt[:,:-1]
    Y = dt[:,-1]
    #normalize the 2,3 columns
    X[:,1] = (X[:,1] - np.mean(X[:,1]))/np.std(X[:,1])
    X[:,2] = (X[:,2] - np.mean(X[:,2]))/np.std(X[:,2])
    #one-hot the 4 column
    X2 = np.zeros((N,D+2))
    X2[:,:-4] = X[:,:-1]
    for i in range(N):
        code = int(X[i,-1])
        X2[i,D-2+code] = 1
    
    return X2, Y
    
```

```python
def get_binary():
    X, Y = get_data()
    X2 = X[Y<2,:]
    Y2 = Y[Y<2]
    
    return X2, Y2
```

```python
X, Y = get_binary()
```
